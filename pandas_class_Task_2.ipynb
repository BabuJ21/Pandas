```python
# 5 . which id is one of the active id that you have in whole dataset 

# Taking the person who has more total steps as active person

fg = pd.DataFrame(df.groupby('Id')['TotalSteps'].mean())
fh = fg.sort_values('TotalSteps', ascending = False)
fh.head(1)
```

```python
# 6 . how many of them have not logged there activity find out in terms of number of ids 

fd = pd.DataFrame(df.groupby(['Id'])['LoggedActivitiesDistance'].sum())
jb = fd[(fd['LoggedActivitiesDistance'] == 0)]
print("The numbers of ids which have not logged there activity are ", len(jb))
jb
```

```python
# 7 . Find out who is the laziest person id that we have in dataset 

fg = pd.DataFrame(df.groupby('Id')['TotalSteps'].mean())
fh = fg.sort_values('TotalSteps', ascending = False)
fh.tail(1)
```

```python
# 8 . Explore over an internet that how much calories burn is required for a healthy person and find out how many healthy person we have in our dataset 

# Most female adults need 1,600–2,200 calories per day, while adult males need 2,200–3,000 calories.
# Taking the average of the calories burn above 1800 as healthy person

fd1 = df[['Id','ActivityDate','Calories']]
fd2 = pd.DataFrame(fd1.groupby('Id')['Calories'].mean())
k = fd2[(fd2['Calories'] > 1800)]
print(k, "\n\n")
print("Number of Healthy person are ", len(k))
```

```python
# 7 . find out how many different differnet shipement mode that we have and what is a percentage usablity of all the shipment mode with respect to dataset 

df1['Ship Mode'].value_counts(normalize=True)*100
```

```python
# 8 . Create a new coulmn and try to find our a diffrence between order date and shipment date

df1['Difference_of_dates'] = df1['Ship Date'].sub(df1['Order Date'], axis=0)
df1['Order_date_day'] = df1['Order Date'].dt.day
df1['Ship_date_day'] = df1['Ship Date'].dt.day
df1['difference_order_ship'] = df1['Ship_date_day']-df1['Order_date_day']

df1['difference'] = df1['Difference_of_dates'].dt.days
df1
```

```python
# 9 . based on question number 8 find out for which order id we have shipment duration more than 10 days 

pd.DataFrame(df1[(df1['difference'] > 10)][['Order ID','difference']])
```

```python
# 10 . Try to find out a list of a returned order which sihpment duration was more then 15 days and find out that region manager as well 

lk = pd.merge(df1, df3, how = 'left', on = 'Region')
lk1 = pd.merge(lk, df2, how = 'left', on = 'Order ID')
lk2 = lk1.iloc[lk1.Status[lk1.Status == 'Returned'].index]

lk2[(lk2['difference'] > 15)][['Order ID','Manager','Status']]
```
